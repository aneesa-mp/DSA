# Week 1 - Project Setup and Data Collection

Key Variables in the Data:
Num_of_Loan: Number of loans the customer has.
Type_of_Loan: Types of loans the customer has (e.g., Auto Loan, Credit-Builder Loan, etc.).
Delay_from_due_date: The delay in payment from the due date.
Num_of_Delayed_Payment: Number of delayed payments.
Changed_Credit_Limit: Change in credit limit.
Num_Credit_Inquiries: Number of credit inquiries.
Credit_Mix: Diversity of credit types.
Outstanding_Debt: Total outstanding debt.
Credit_Utilization_Ratio: Ratio of credit used to total available credit.
Credit_History_Age: Age of the credit history.
Payment_of_Min_Amount: Indicates if the minimum amount is paid or not.
Total_EMI_per_month: Total EMI paid per month.
Amount_invested_monthly: Amount invested monthly.
Payment_Behaviour: Payment behavior (e.g., High_spent_Small_value_payments, Low_spent_Large_value_payments, etc.).
Monthly_Balance: Monthly balance.
Credit_Score: Credit score category (Good, Standard, Poor).
